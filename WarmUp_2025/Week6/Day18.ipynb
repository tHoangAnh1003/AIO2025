```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Bài 1:

```python
df1 = pd.read_csv('./student.csv')
df2 = pd.read_csv('./extra.csv')
```

```python
df2['id'] = range(0, len(df2))
df2.set_index('id', inplace=True)
```

```python
pd.merge(df1, df2, on='id', how='inner').head()
```

Bài 2:

```python
df1.pivot_table(values='math.grade', index='nationality', columns='gender', aggfunc='sum')
```

```python
df1.pivot_table(values='english.grade', index='ethnic.group', columns='gender', aggfunc='mean')
```

Bài 3:

```python
avg_science = df1.groupby('gender')['sciences.grade'].mean()
avg_science.plot(kind='bar', title='Average Sciences Grade by Gender', ylabel='Average Grade', xlabel='Gender', color=['skyblue', 'salmon'])

plt.tight_layout()
plt.show()
```

```python
df1['age'].hist()
```

```python
corr_matrix = df1[['english.grade', 'math.grade', 'sciences.grade', 'language.grade']].corr()

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Heatmap of Grade Correlation Matrix')
plt.show()
```
